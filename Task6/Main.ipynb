# Задача 1: Двумерный численный интеграл

```python
def func(x, y):
    return x**2 + y**2 + 1

x = np.linspace(0, 1.2, 13)
y = np.linspace(0, 1.2, 13)
z = [[func(x[i], y[j]) for i in range(len(x))] for j in range(len(y))]

integrals = []
N_Y = 20
N_X = 10
f = Spline2DFunc(x, y, z)

ySurf = np.linspace(0, 1, N_Y)
for i in range(len(ySurf)):
    fRow = lambda x: f(x, ySurf[i])
    integrals.append(GaussIntegral(0, 1 - ySurf[i], N_X, fRow))

f = SplineFunc(ySurf, integrals)

res = SympsonIntegral(y, f)
print(f"Двукратный интеграл: {res}")
```
